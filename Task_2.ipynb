```python
#1)Unique Words in a Sentence

A="python is easy and python is powerful"
# sent=input('Enter sentence')
b=set(A.split())
print('unique counts: ',len(b))
print('unique words: ',(b))
```

```python
#2)Highest Salary from Employee Data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest = max(employees, key=employees.get)
print("Highest Salary:", highest, "-", employees[highest])
```

```python
#3)Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]
print('Max:',max(numbers))
print('Min:',min(numbers))
```

```python
# 4) Count Products Above a Price Threshold

prices = [450, 1200, 899, 1500, 300]
count=0

for price in prices:
    if price > 1000:
        count += 1
print("Products above 1000:", count)
```

```python
#5)Calculate Attendance Percentage


attendance = ["P", "P", "A", "P", "P"]
count=0
for ch in attendance:
  if ch == 'P':
    count +=1
print((count/5)*100)
```

```python
#6)Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 1234567890, 9876543210]
print(set(phone_numbers))
```

```python
#7)Count Character Frequency

text = "python"
freq = {}

for ch in text:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

print(freq)
```

```python
#9)Check if a Key Exists in a Dictionary

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
key = "Ravi"

if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```
