```python
# Problem 1: Social Media – Post Engagement Analyzer
def analyze_engagement(likes):
    total = 0

    for like in likes:
        total += like

    if total >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total)
    print("Post Status:", status)

likes = [200, 300, 150, 400]

analyze_engagement(likes)
```

```python
# Problem 2: Healthcare – Medicine Stock Alert System
def check_medicine_stock(stock):
    print("Medicine Stock:", stock)

    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")

stock = 6

check_medicine_stock(stock)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker
def check_rainfall(rainfall, required_level):
    total = 0

    for value in rainfall:
        total += value

    average = total / len(rainfall)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)

rainfall = [70, 75, 80, 65, 70]
required_level = 70

check_rainfall(rainfall, required_level)
```

```python
# Problem 4: Social Media – Duplicate Account Detection
def detect_duplicate_accounts(usernames):
    unique = set(usernames)

    if len(unique) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

usernames = ["user1", "user2", "user3", "user1"]

detect_duplicate_accounts(usernames)
```

```python
# Problem 5: Healthcare – Appointment Eligibility Checker
def check_appointment(age):
    print("Patient Age:", age)

    if age >= 18:
        print("Eligibility Status: Eligible")
    else:
        print("Eligibility Status: Not Eligible")

age = 21

check_appointment(age)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter
def filter_premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)

prices = [1500, 2500, 1800, 3200]

filter_premium_crops(prices)
```

```python
# Problem 7: System Monitoring – Application Health checker
def check_application_health(errors):
    print("Error Count:", errors)

    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")

errors = 7

check_application_health(errors)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker
def check_transaction(amount):
    print("Transaction Amount:", amount)

    if amount <= 50000:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")

amount = 60000

check_transaction(amount)
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System
def check_exam_eligibility(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance = ["P", "P", "A", "P", "P"]

check_exam_eligibility(attendance)
```

```python
# Problem 10: Smart Electricity Bill Analyzer
def calculate_electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage

units_consumed = 250
bill, usage_status = calculate_electricity_bill(units_consumed)

print("Total Bill:", bill)
print("Usage Classification:", usage_status)
```
