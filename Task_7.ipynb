```python
#1) Smart Light Controller Create a class to control a smart light that can be turned ON or OFF and display its status.


class SmartLight:
    def __init__(self, name):
        self.name = name
        self.is_on = False

    def control(self, action):
        if action.upper() == "ON":
            self.is_on = True
        elif action.upper() == "OFF":
            self.is_on = False
        else:
            return f"Invalid action for {self.name}"

        status = "ON" if self.is_on else "OFF"
        return f"{self.name} is {status}"


# Sample input matching your test
light = SmartLight("Bedroom Light")
print(light.control("ON"))  # Bedroom Light is ON
print(light.control("OFF"))  # Bedroom Light is OFF
print(light.control("TOGGLE"))  # Invalid action for Bedroom Light

```

```python
#2) Employee ID Card System Create a class that stores employee name, ID, and department and displays ID card details.


class EmployeeIDCard:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")



emp = EmployeeIDCard("Rahul", "EMP102", "IT")
emp.display()


```

```python
#3)Mobile Contact Record Create a class to store contact name and phone number and display contact information.


class MobileContact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def save_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")



# Matches your sample input exactly
contact = MobileContact("Anita", "9876543210")
contact.save_contact()
```

```python
 #4) Product Price Tag Generator Create a class that stores product name and price and prints a formatted price tag.


class ProductPriceTag:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")



# Sample input test
product = ProductPriceTag("Headphones", 2499)
product.display_tag()
```

```python
 #5) Movie Rating Display System Create a class that stores movie name and rating and displays movie details.


class MovieRating:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")



# Sample input test
movie = MovieRating("Inception", 4.8)
movie.display()
```

```python
 #6) Delivery Address Manager Create a class that stores customer name and delivery address and prints delivery details.


class DeliveryAddress:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")



# Sample input test
delivery = DeliveryAddress("Hema", "Hyderabad")
delivery.display_details()
```
