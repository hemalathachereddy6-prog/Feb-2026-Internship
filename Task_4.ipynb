```python
Problem 1: Mobile Recharge Validation System
```

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    # Rule 1: Minimum amount
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False

    # Rule 2: Valid plan check
    if amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans: 199, 299, 399, 599")
        return False

    return True


# Retry system using while loop
while True:
    amount = int(input("Enter recharge amount: ₹"))

    if validate_recharge(amount):
        print("Recharge successful!")
        break
    else:
        print("Please try again.\n")
```

```python
Problem 2: Inventory Reorder Alert System
```

```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(product, "-> Reorder Alert (Stock:", stock, ")")
        else:
            print(product, "-> Stock OK (Stock:", stock, ")")


# Sample dictionary
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18
}

check_inventory(inventory)
```

```python
Problem 3: Student Result Processing System
```

```python
def process_result(marks):
    total = 0

    # Calculate total using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Example usage
student_marks = [60, 45, 70, 55]

result = process_result(student_marks)

print("Average Marks:", sum(student_marks) / len(student_marks))
print("Result:", result)
```

```python
Problem 4: Cab Fare Estimator with Retry Option
```

```python
def calculate_fare(distance, is_peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    if is_peak:
        fare = fare + (fare * 0.25)   # 25% extra

    return fare

# Retry system
while True:
    distance = float(input("Enter distance in km: "))
    peak_input = input("Is it peak hour? (yes/no): ").lower()

    if peak_input == "yes":
        is_peak = True
    else:
        is_peak = False

    total_fare = calculate_fare(distance, is_peak)

    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        print("Thank you for using Cab Fare Estimator.")
        break
```

```python
Problem 5: Employee Attendance Eligibility Checker
```

```python
def check_attendance(attendance_list):
    present_days = 0

    # Count present days using loop
    for day in attendance_list:
        if day == "P":
            present_days += 1

    total_days = len(attendance_list)
    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example usage
attendance = ["P", "A", "P", "P", "P", "A", "P", "P", "P"]

result = check_attendance(attendance)

print("Attendance Percentage:", (attendance.count("P") / len(attendance)) * 100)
print("Eligibility Status:", result)
```

```python
Problem 6: Password Strength Checker
```

```python
def check_password_strength(password):
    has_digit = False
    has_special = False

    # Rule 1: Check minimum length
    if len(password) < 8:
        return "Weak Password (Minimum length is 8)"

    # Loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True

        if char in "@#$":
            has_special = True

    # Final validation
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character)"


# Example usage
user_password = "Shital@111"

result = check_password_strength(user_password)

print("Password Status:", result)
```
